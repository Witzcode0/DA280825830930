```python
data = {
    "name": ["A", "B", "C", "D", "E", "F"],
    "marks": [97, 68, 89, 78, 90, 58]
}

df = pd.DataFrame(data)
df
```

```python
def get_grade(mark):
    if mark >= 90:
        return "AA"
    elif mark >= 80:
        return "BB"
    else:
        return "CC"

df["grade"] = df["marks"].map(get_grade)
df
```
