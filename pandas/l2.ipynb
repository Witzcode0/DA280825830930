```python
data = {
    "name": ["Bob", "Aman", "Jay", "Raj", "Nayan", "Neha", "Bob"],
    "marks" : [89, 78, None, 92, 75, np.nan, 89],
    "attendance":  [95, 88, 56, None, 78, np.nan, 95]
}

df = pd.DataFrame(data)
df
```

```python
df.isnull()
df.isnull().sum()
df.notnull()
df.dropna()
```

```python
df["marks"] = df["marks"].fillna(df["marks"].mean())
df["attendance"] = df["attendance"].fillna(100)
df
```
