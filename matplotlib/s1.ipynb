```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
chocolates = [2, 4, 6, 3, 7, 8, 10]

# plt.plot(days, chocolates, "r>-", linestyle=":", color="#32a896") # Line style: -, --, -., :
# plt.plot(days, chocolates, "r>-") # Line style: -, --, -., :
plt.plot(days, chocolates, "b*-") # Line style: -, --, -., :

plt.title("Chocolate I eat each day")

plt.xlabel("Days")
plt.ylabel("numbers")

plt.savefig("chocolate_eat_each_day.png")

plt.show()
```

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
# Small blurry image
plt.figure(figsize=(3, 2), dpi=50)
plt.plot(x, y,'r*-')
plt.title("Small & Blurry (3x2 inches, 50 DPI)")
plt.show()

plt.figure(figsize=(6, 4), dpi=100)
plt.plot(x, y,
'b*-')
plt.grid(True)
plt.title("Large & Sharp (6x4 inches, 100 DPI)")
plt.show()

```

```python
years = [2018, 2019, 2020, 2021, 2022]
height = [120, 130, 140, 150, 160]

plt.plot(years, height, "ro-", label="my line 1")
plt.plot(years, [h+5 for h in height], "g^-", label="my line 2")

plt.legend()

plt.show()
```
