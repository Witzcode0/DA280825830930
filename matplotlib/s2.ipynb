```python
import matplotlib.pyplot as plt
```

```python
# Line graph

days = [1,2,3,4,5]
temp = [30, 32, 31, 33, 35]

plt.plot(days, temp, "g-o", label="Temprature")

plt.xlabel("Days")
plt.ylabel("Temp (C)")

plt.grid(True)

plt.legend()

plt.show()
```

```python
students = ["A", "B", "C", "D"]
grade = [85, 70, 90, 60]

plt.bar(students, grade, color="#00ff00")

plt.title("Students Marks(vertical)")
plt.ylabel("Marks")

plt.show()
```

```python
students = ["A", "B", "C", "D"]
grade = [85, 70, 90, 60]

plt.barh(students, grade, color="#0000ff")

plt.title("Students Marks(Horizontal)")
plt.ylabel("Marks")

plt.show()
```

```python
score = [45, 55, 60, 70, 72, 85, 90, 95, 100, 66, 77, 88, 45, 59, 62]

plt.hist(score, bins=5, color="#4a7d9c", edgecolor="#00ff00")

plt.xlabel("score range")

plt.ylabel("Number of students")

plt.show()
```

```python
hours = [2,3,4,5,6,7,8]
marks = [50, 55, 60, 65, 70, 80, 85]

plt.scatter(hours, marks, color="#d6c033", marker="^")

plt.show()
```

```python
expences = [3000, 1500, 2000, 1000, 1200]
categories = ["Food", "Travel", "Rent", "Other", "ele"]

plt.pie(expences, labels=categories, autopct="%1.1f%%", colors= ["#62109F", "#DC0E0E","#FE6244","#FFDEB9", "#87BAC3"])

plt.show()
```

```python
months = ["jun", "Jul", "Aug", "Sept", "Oct"]
rainfall = [20, 40, 34, 50, 60]

plt.fill_between(months, rainfall, color="#79a2ba", alpha=0.1)

plt.grid()

plt.plot(months, rainfall, linestyle="--", color="blue", marker="^")

plt.show()
```

```python
days = [1,2,3,4,5,6,7]

kitchen = [3,4,3,5,4,6,5]
heating = [5,6,7,6,5,4,5]
lights = [2,2,3,3,2,2,3]

plt.stackplot(days, kitchen, heating, lights, labels=["Kitchen", "heating", "Lights"], colors=["#ff0000", "#00ff00", "#0000ff"])
plt.legend(loc="upper left")
plt.show()
```
